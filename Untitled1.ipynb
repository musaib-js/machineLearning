```python
import tensorflow as tf
import logging

tf.get_logger().setLevel(logging.ERROR)

import numpy as np
```

```python
# Setting up the training data
celcius_i = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float)
farenheit_o = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)

for i, c in enumerate(celcius_i):
  print("{} degree Celcius = {} degree Farenheit".format(c, farenheit_o[i]))
```

```python
l0 = tf.keras.layers.Dense(units = 1, input_shape = [1] )
```

```python
model  = tf.keras.Sequential([l0])
```

```python
model.compile(loss = 'mean_squared_error', optimizer  = tf.keras.optimizers.Adam(0.1))
```

```python
history = model.fit(celcius_i, farenheit_o, epochs=500, verbose = False)
print('Finished training the model')
```

```python
import matplotlib.pyplot as plt
plt.xlabel('Epoch Number')
plt.ylabel('Loss Magnitude')
plt.plot(history.history['loss'])
```

```python
print(model.predict([123]))
```
